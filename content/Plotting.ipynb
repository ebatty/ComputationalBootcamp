## Styles of plotting

There are two "styles" for plotting in matplotlib, outlined below.

### Style 1: Functional programming aka MATLAB-style

In this style, you use a bunch of commands to make and customize your plots.

```python
plt.plot(np.arange(3))
plt.xlabel('x label')
plt.ylabel('y label')
```

```python
plt.subplot(1, 2, 1)
plt.plot(np.arange(5))

plt.subplot(1, 2, 2)
plt.plot(np.arange(5))
```

### Style 2: Object Oriented Interface

In this style, the axes of different subplots are objects whose attributes you can change. This means you can easily reference subplots out of order, and it takes less code to set some of the customization.

```python
fig, ax = plt.subplots()

ax.plot(np.arange(5))

#ax.set_xlabel('x label')
ax.set(xlabel = 'x label', 
       ylabel = 'y label');
```

```python
fig, axes = plt.subplots(1, 2)

axes[0].plot(np.arange(3))
axes[1].plot(np.arange(5))

axes[0].set(title='Hello')
```

## Plotting basics

See below for how to add axis labels, a title, where the ticks on x/y axes should occur, how to customize the tick labels, and change the axis limits. Using the second style from above, you can do all of this with `set` instead of needing separate commands for all.

```python
fig, ax = plt.subplots()

data = np.arange(3)
ax.plot([0, 1, 2], data)

ax.set(xlabel = 'x', 
       ylabel = 'y', 
       title = 'title', 
       xticks = [.5, 1.5], 
       xticklabels = ['Monday', 'Friday'],
       ylim = [0, .2]);
```

### Changing line properties

```python
fig, ax = plt.subplots()

data = np.arange(3)

# You can specify all of the customization of line properties with separate arguments for each one
#ax.plot([0, 1, 2], data, linestyle = '--', marker = 'o', color = 'r', linewidth = 3)

# You can also use a shortcut where you can specify line style, marker style, and color easily
ax.plot(data, '--or')

# You can also use any color by specifying hex codes
# ax.plot(data, '-o', color = '#1EC2A0')
```

### Add legend

```python
# Option 1: specify after plotting

fig, ax = plt.subplots()

ax.plot(data, '-ok')
ax.plot(data*2, '-og')

ax.legend(['Label 1', 'Label 2'])
```

```python
# Option 2: specify while plotting
fig, ax = plt.subplots()

ax.plot(data, '-ok', label = 'Label 1')
ax.plot(data*2, '-og', label = 'Label 2')

ax.legend()
```

### Change figure size

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax.plot(data)
```

### Change font size

```python
# This will change all font size for this and future plots
import matplotlib
matplotlib.rcParams['font.size'] = 20

fig, ax = plt.subplots()
ax.plot(data)
ax.set_xlabel('x')
```

## Other types of plots

### Scatter plot

You can make something that looks like a scatter plot with `ax.plot` but all the markers have to be the same size, color, etc. With `ax.scatter`, you can customize individual data points.

```python
fig, ax = plt.subplots()
ax.scatter([0, 1, 2], [5, 3, 4], c = ['g', 'b', 'k'])
```

### Images

```python
x = np.random.randn(10, 10)

fig, ax = plt.subplots()
im = ax.imshow(x, cmap = 'gray')
plt.colorbar(im)
```

### Histograms

```python
fig, ax = plt.subplots()
ax.hist(np.random.randn(100, 10).reshape((-1,)), bins = 100);
```

### Bar Plots

```python
fig, ax = plt.subplots()
ax.bar([0, 1, 2], [10, 4, 1], yerr = [1, 1, 2])
```

### Line plot with error bars

```python
fig, ax = plt.subplots()
ax.errorbar([0, 1, 2], [10, 3, 6], yerr = [1, 1, 2])
```

```python
fig, ax = plt.subplots()
ax.plot([0, 1, 2], [10, 3, 6])
y = np.array([10, 3, 6])
yerr = np.array([1, 1, 2])
ax.fill_between([0, 1, 2], y - yerr, y + yerr, alpha = .2)
```

## Subplots

### Standard subplots

```python
fig, axes = plt.subplots(nrows = 3, ncols = 2, figsize = (10, 5))

axes[0][1].set_title('hello')

plt.tight_layout()
```

If subplot layout still doesn't look nice with tight_layout, try changing the proportions of the fig size to suit the plots

How do you create 6 rows and 4 columns of subplots and ensure that they're all square?

```python
fig, ax = plt.subplots(6, 4, figsize = (4, 6))
```

### Gridspec

GridSpec is useful for more complicated subplot arrangements

```python
fig = plt.figure()

grid = fig.add_gridspec(3, 3)

ax1 = fig.add_subplot(grid[:, 0])

ax2 = fig.add_subplot(grid[0, 1:])
```
